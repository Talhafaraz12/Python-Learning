```python
Details_of_Banks = pd.read_csv('Monthly Capital of Top Banks.csv' , encoding = 'latin-1' )
Details_of_Banks.head()
```

```python
month_columns = Details_of_Banks.columns[1:] # Jan through Dec

def clean_data(value):
    if isinstance(value, str):
        # Keep only numbers and decimal points
        value = ''.join(c for c in value if c.isdigit() or c == '.')
    return float(value)
```

```python
for col in month_columns:
    Details_of_Banks[col] = Details_of_Banks[col].apply(clean_data)
```

```python
X = np.array(range(1, 13)).reshape(-1, 1) 
X_future = np.array(range(13, 16)).reshape(-1, 1)

predictions_list = []
```

```python
# STEP 4: Train and Predict for each bank
for index, row in Details_of_Banks.iterrows():
    bank_name = row['Bank Name']
    y = row[month_columns].values # The 12 months of capital
    
    # Initialize the model
    model = LinearRegression()
    
    # "Fit" the model (This is the actual Machine Learning part)
    model.fit(X, y)
    
    # Predict the future
    future_capital = model.predict(X_future)
    
    # Store the results
    predictions_list.append({
        'Bank': bank_name,
        'Janauray 26': round(future_capital[0], 2),
        'Febuary 26': round(future_capital[1], 2),
        'March 26': round(future_capital[2], 2),
        'Trend_Strength (R2)': round(model.score(X, y), 3)
    })
```

```python
# STEP 5: View Results
results_df = pd.DataFrame(predictions_list)
print(results_df)

# Save it to a new file
results_df.to_csv('my_bank_predictions.csv', index=False)
```
